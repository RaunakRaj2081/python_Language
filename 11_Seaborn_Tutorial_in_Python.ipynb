#Seaborn

✅ What is Seaborn?

Seaborn is a Python library used for data visualization.

It is built on top of Matplotlib, but it is easier to use and looks better by default.

It works best with Pandas DataFrames and helps you quickly explore and visualize relationships in data.

#Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

✅ sns.relplot()

It stands for "relational plot" and is used to plot relationships between two numeric columns.

By default, it uses a scatter plot.

✅ data=tips

tips is a built-in dataset in Seaborn that contains restaurant bill and tip data.

Each row is a record of a bill with details like total bill, tip, gender, smoking status, etc.

✅ x='total_bill', y='tip'

X-axis will show total_bill (how much the customer spent).

Y-axis will show tip (how much tip they gave).

✅ col='time'

This creates separate plots (columns) for each value in the time column (usually Lunch and Dinner).

So you’ll get two plots side by side: one for Lunch, one for Dinner.

✅ hue='smoker'

Adds color based on whether the customer is a smoker or non-smoker.

So, different colors show smokers and non-smokers.

✅ style='smoker'

Changes the marker shape depending on smoking status (e.g., circle for non-smoker, triangle for smoker).

✅ size='size'

Changes the size of each point based on the number of people at the table.



📊 Final Output:

You get two scatter plots side by side (Lunch and Dinner), where:

Color and shape represent smoker vs non-smoker.

Size of dots shows how many people were at the table.

X and Y axes show total bill and tip.



```python
#visulalize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

#Heat Map

plt.figure(figsize=(10,10))

This sets the size of the figure to 10x10 inches.

Makes sure the heatmap is big enough to read clearly.

Now let’s break each argument:

✅ correlation

This should be a correlation matrix, usually created like this:

correlation = data.corr()
It shows how one variable is related to another. Values range from:

+1 → strong positive relation

-1 → strong negative relation

0 → no relation

✅ cbar=True ->
Shows the color bar on the side, which helps you understand the color scale.

✅ square=True ->
   Makes each cell in the heatmap a perfect square (neat layout).

✅ fmt='.1f' ->
Formats the numbers inside the boxes to 1 decimal place (like 0.8, -0.5).

✅ annot=True ->
Turns on annotations so each square shows its numeric value.

✅ annot_kws={'size':8} ->
Sets the font size of the annotation text to 8.

✅ cmap='Blues' ->
Sets the color scheme to different shades of blue. Darker = higher correlation.





```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation , cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

#📊 Dataset Example

```python
df = sns.load_dataset('tips')  # Loads tips dataset
print(df.head())
```

#🌈 1. Basic Plot: sns.histplot()

```python
sns.histplot(data=df, x='total_bill')
plt.show()
#📌 This creates a histogram of the total_bill column.
```

#📈 2. Line Plot: sns.lineplot()

```python
sns.lineplot(data=df, x='size', y='tip')
plt.show()
#Shows a line graph of group size vs. tip amount.
```

#🔲 3. Scatter Plot: sns.scatterplot()

```python
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.show()
#Shows the relationship between total_bill and tip.
```

#🧱 4. Bar Plot: sns.barplot()

```python
sns.barplot(data=df, x='sex', y='tip')
plt.show()
#Compares the average tip given by males and females.
```

#📏 5. Box Plot: sns.boxplot()

```python
sns.boxplot(data=df, x='day', y='total_bill')
plt.show()
#Shows spread of data and outliers for each day’s total bill.
```

#🎯 6. Violin Plot: sns.violinplot()

```python
sns.violinplot(data=df, x='day', y='tip')
plt.show()
#Like a box plot, but also shows the distribution shape.
```

#📚 7. Count Plot: sns.countplot()

```python
sns.countplot(data=df, x='day')
plt.show()
#Shows how many times each category appears.
```

#🧠 8.Pair Plot: sns.pairplot()

```python
sns.pairplot(df)
plt.show()
#Creates scatter plots between all pairs of numeric columns.
#Great for quick data analysis.
```
