#Matplotlib



ðŸ§  What is Matplotlib?

Matplotlib is a Python library used for creating visualizations and graphs like:

Line graphs

Bar charts

Pie charts

Histograms

Scatter plots

It helps to see patterns in data clearly.

âœ… Why Use Matplotlib?

Shows data visually (easier to understand).

Supports many types of charts.

Easy to customize graphs (colors, labels, titles).

Works well with NumPy and Pandas.

Useful in data science and machine learning.



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
#Plotting a Simple Line Graph
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.show()

#plt.plot() â†’ creates the line graph.
#plt.show() â†’ displays the graph.
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

#Plotting the data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

# Adding title, x-axis & y-axis labels

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

#Change Line Style and Color

color='green' â†’ line color

linestyle='--' â†’ dashed line

marker='o' â†’ dots at each point

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title("Styled sine wave")
plt.show()
```

#Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  #[left, bottom, width, height]
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.title('Bar Chart')
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

ðŸ” What [0, 0, 1, 1] Means

0: Start from the left edge (0%)

0: Start from the bottom edge (0%)

1: Width is 100% of the figure

1: Height is 100% of the figure

ðŸŸ¢ So this means:

"Create a graph area that fills the entire figure."

#Pie Chart

 autopct='%1.1f%%'

This tells the chart to show percentage values on each slice.

'%1.1f%%' means:

1.1f â†’ show 1 digit before and 1 digit after the decimal (like 25.0%)

%% â†’ the percent symbol (escaped because % is a special character)

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

#Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

#3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100) # This generates 100 random numbers between 0 and 20.
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')  # c=z â†’ the color of each point depends on the z-value (height).
plt.show()
```

#Histogram

```python
data = [22, 30, 26, 35, 40, 41, 45, 48, 50, 55, 60, 65, 70]

plt.hist(data, bins=5, color='skyblue')
# bins = 5 -> "Divide all my data into 5 groups (ranges) and show how many data points fall into each group."
plt.title("Histogram")
plt.show()
```

#Save Graph as Image File

```python
x = [2,3,4,5]
y = [5,7,9,8]

plt.plot(x, y)
plt.savefig("my_plot.png")  # Save as image
```

#Customize Fonts and Colors

```python
plt.plot(x, y, color='purple')
plt.title("Custom Font", fontsize=14)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True)  # Add gridlines
plt.show()
```

#Subplots (Multiple Graphs in One Window)

```python
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot This tells Matplotlib: "Prepare a layout of 1 row and 2 columns, and now plot in the 1st position."
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Plot 1")

plt.subplot(1, 2, 2)  # 2nd plot
plt.plot([1, 2, 3], [6, 5, 4])
plt.title("Plot 2")

plt.tight_layout()  # Avoid overlapping
plt.show()
```
